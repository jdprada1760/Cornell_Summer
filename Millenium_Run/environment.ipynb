```python
table = array(loadtxt("temp.data", skiprows = 0, delimiter = ',').T)
#print table
```

```python
#POV
r0 = 250.0*np.ones((1,3))
# Halo position and velocity
rh = table[0:3].T
vh = table[3:6].T
# Galaxies positions and velocities
rg = table[6:9].T
vg = table[9:12].T
# peculiar velocity difference and position difference
dv = vg - vh
dr = rg - rh
# the total velocity difference according to the hubbles law
dvt = dv + H*dr
# distance to the halo
d = (rh-r0)[0]/np.linalg.norm((rh-r0)[0])
# Hubbles constant
H = 100.0
d,np.linalg.norm(d)
```

```python
#indexes were the cut in velocity is satisfied
ies = []
for i in range(len(dvt)):
    # obtain the velocity projection according the unit vector the pov to the halo
    if( np.dot(d,dvt[i]) < 500 ):
        ies.append(i)
ies = array(ies)
len(ies)
```

```python
# Calculate the projected distance of galaxies satisfying the velocity cut
distances = []
for i in ies:
    distances.append(np.linalg.norm((rg[i] - r0) - d*np.dot(rg[i] - r0, d)))
distances = array(distances)
```
