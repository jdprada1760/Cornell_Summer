```python
# Gives the x,y of the histogram for logmasses from the given array of mass
def histog( mf, nbins ):
    hist, bin_edges = np.histogram(mf, bins = nbins)
    # Obtains the center point for each bin
    xcenter = (bin_edges[:-1] + bin_edges[1:])/2
    return np.array(xcenter), np.array(hist/vol)
```

```python
# To avoid log10(0)
def log100(x):
    return log10(x)
```

```python
# Schechter function
def Schechter(m,alpha,phi_s,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))*numpy.exp(-MHI/M_s)
# Power fit
def Power(m,alpha,phi_s,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))
# Returns the fitted parameterts for the Schechter function given a mass distribution
def fit(mf,bins):
    # tmp[0]:mj,  tmp[1]:nj
    tmp = histog(log100(mf),bins)
    return optimize.curve_fit(Schechter,tmp[0] ,tmp[1],p0=[-2.0,0,15], sigma = np.sqrt(tmp[1]*vol)/vol, maxfev = 100000)
# Returns the fitted parameters for Power law given the mass distribution
def fitP(mf,bins):
    tmp = histog(log100(mf),bins)
    return optimize.curve_fit(Power,tmp[0],tmp[1],p0=[-2.0,0,14], sigma = np.sqrt(tmp[1]*vol)/vol, maxfev = 40000)
```

```python
# loads the file with halo masses and makes the quartiles
# i: position of the distance, j: position of the mass
def loadHaloes(name, neigh, i, j):
    table = np.loadtxt(name, skiprows = 0, delimiter = ',').T
    table[i] = 3/(np.pi*table[i])
    print "number of haloes: " + str(len(table.T))
    # Calculates the first quartile of logsigma
    global q1
    global q2
    global q3
    global q4
    global q5
    lsigma = table[i]
    mass = M*table[j]
    q = np.percentile(lsigma,25)
    # Keeps first quartile
    q1 = mass[where(lsigma <= q)]
    # Removes fist quartile from lsigma
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Second quartile
    q = np.percentile(lsigma,33.333333)
    q2 = mass[where(lsigma <= q)]
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Third and Fourth quartile
    q = np.percentile(lsigma,50)
    q3 = mass[where(lsigma <= q)]
    q4 = mass[where(lsigma >= q)]
    q5 = M*table[j]
    # verify the length of each quartile
    print len(q1),len(q2),len(q3),len(q4), sum([len(q1),len(q2),len(q3),len(q4)])
    
def SchFitGraphs(vol, bins, name):
    # Graphics of the mass function for each quartile
    import matplotlib.gridspec as gridspec
    fig, axes = plt.subplots( nrows=2, ncols=4, figsize=(15,20) )
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.3)
    gs = gridspec.GridSpec(4,2)
    ax0 = plt.subplot(gs[0, 0])
    ax1 = plt.subplot(gs[0, 1])
    ax2 = plt.subplot(gs[1, 0])
    ax3 = plt.subplot(gs[1, 1])
    ax4 = plt.subplot(gs[2:,:])
    # Histo for first quartile
    tmp = histog(log100(q1),bins)
    array([tmp, sqrt(vol*tmp[1])/(vol*tmp[1])]).tofile(name + "q1.csv", sep = ',')
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax0.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b',
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fit(q1,bins)[0]
    errors = fit(q1,bins)[1]
    print ft[1],errors[1][1]
    ax0.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r', 
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(errors[2][2])))
    ax0.set_title('First Quartile', fontsize = 18)
    ax0.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax0.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax0.set_ylim(min(log100(tmp[1]))-1,0)
    ax0.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax0.legend(prop={'size':10})
    #--
    tmp = histog(log100(q2),bins)
    array([tmp, sqrt(vol*tmp[1])/(vol*tmp[1])]).tofile(name + "q2.csv", sep = ',')
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax1.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fit(q2,bins)[0]
    errors = fit(q2,bins)[1]
    print ft[1],errors[1][1]
    ax1.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(errors[2][2])))
    ax1.set_title('Second Quartile', fontsize = 18)
    ax1.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax1.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax1.set_ylim(min(log100(tmp[1]))-1,0)
    ax1.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax1.legend(prop={'size':10})
    #--
    tmp = histog(log100(q3),bins)
    array([tmp, sqrt(vol*tmp[1])/(vol*tmp[1])]).tofile(name + "q3.csv", sep = ',')
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax2.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fit(q3,bins)[0]
    errors = fit(q3,bins)[1]
    print ft[1],errors[1][1]
    ax2.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(errors[2][2])))
    ax2.set_title('Third Quartile', fontsize = 18)
    ax2.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax2.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax2.set_ylim(min(log100(tmp[1]))-1,0)
    ax2.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax2.legend(prop={'size':10})
    #--
    tmp = histog(log100(q4),bins)
    array([tmp, sqrt(vol*tmp[1])/(vol*tmp[1])]).tofile(name + "q4.csv", sep = ',')
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax3.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fit(q4,bins)[0]
    errors = fit(q4,bins)[1]
    print ft[1],errors[1][1]
    ax3.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(errors[2][2])))
    ax3.set_title('Fourth Quartile', fontsize = 18)
    ax3.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax3.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax3.set_ylim(min(log100(tmp[1]))-1,0)
    ax3.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax3.legend(prop={'size':10})

    # Total fit
    bins = 25
    tmp = histog(log100(q5),bins)
    array([tmp, sqrt(vol*tmp[1])/(vol*tmp[1])]).tofile(name + "total.csv", sep = ',')
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax4.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b',
                 c = 'b', markersize=4,elinewidth=1.5)
    ft = fit(q5,bins)[0]
    errors = fit(q5,bins)[1]
    print ft[1],errors[1][1]
    ax4.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r', 
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(errors[2][2])))
    ax4.set_title('Total', fontsize = 18)
    ax4.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax4.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax4.set_ylim(min(log100(tmp[1]))-1,0)
    ax4.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax4.legend(prop={'size':10})

def PowFitGraphs(vol, bins):
    # Graphics of the mass function for each quartile
    import matplotlib.gridspec as gridspec
    fig, axes = plt.subplots( nrows=2, ncols=4, figsize=(15,20) )
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.3)
    gs = gridspec.GridSpec(4,2)
    ax0 = plt.subplot(gs[0, 0])
    ax1 = plt.subplot(gs[0, 1])
    ax2 = plt.subplot(gs[1, 0])
    ax3 = plt.subplot(gs[1, 1])
    ax4 = plt.subplot(gs[2:,:])
    # Histo for first quartile
    tmp = histog(log100(q1),bins)
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax0.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b',
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fitP(q1,bins)[0]
    errors = fitP(q1,bins)[1]
    ax0.plot(logm, log100(Power(logm,ft[0],ft[1],ft[2]) ),'r', 
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(abs(errors[2][2]))))
    ax0.set_title('First Quartile', fontsize = 18)
    ax0.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax0.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax0.set_ylim(min(log100(tmp[1]))-1,0)
    ax0.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax0.legend(prop={'size':10})
    #--
    tmp = histog(log100(q2),bins)
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax1.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fitP(q2,bins)[0]
    errors = fitP(q2,bins)[1]
    ax1.plot(logm, log100(Power(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(abs(errors[2][2]))))
    ax1.set_title('Second Quartile', fontsize = 18)
    ax1.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax1.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax1.set_ylim(min(log100(tmp[1]))-1,0)
    ax1.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax1.legend(prop={'size':10})
    #--
    tmp = histog(log100(q3),bins)
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax2.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fitP(q3,bins)[0]
    errors = fitP(q3,bins)[1]
    ax2.plot(logm, log100(Power(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(abs(errors[2][2]))))
    ax2.set_title('Third Quartile', fontsize = 18)
    ax2.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax2.set_xlabel('$L_{10}(M)$', fontsize = 18)
    ax2.set_ylim(min(log100(tmp[1]))-1,0)
    ax2.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax2.legend(prop={'size':10})
    #--
    tmp = histog(log100(q4),bins)
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax3.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b', 
                 c = 'b', markersize=2.5,elinewidth=1.5)
    ft = fitP(q4,bins)[0]
    errors = fitP(q4,bins)[1]
    ax3.plot(logm, log100(Power(logm,ft[0],ft[1],ft[2]) ),'r',
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(abs(errors[2][2]))))
    ax3.set_title('Fourth Quartile', fontsize = 18)
    ax3.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax3.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax3.set_ylim(min(log100(tmp[1]))-1,0)
    ax3.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax3.legend(prop={'size':10})

    # Total fit
    bins = 25
    tmp = histog(log100(q5),bins)
    logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
    ax4.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'b',
                 c = 'b', markersize=4,elinewidth=1.5)
    ft = fitP(q5,bins)[0]
    errors = fitP(q5,bins)[1]
    ax4.plot(logm, log100(Power(logm,ft[0],ft[1],ft[2]) ),'r', 
             label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"d$\alpha$ = " + str(sqrt(errors[0][0]))+"\n"
             + r"$m_\ast$ = " + str(ft[2])  +"\n" + r"d$m_\ast$ = " + str(sqrt(abs(errors[2][2]))))
    ax4.set_title('Total', fontsize = 18)
    ax4.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax4.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax4.set_ylim(min(log100(tmp[1]))-1,0)
    ax4.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
    ax4.legend(prop={'size':10})
    
def histo(vol, bins):
    # Graphics of the mass function for each quartile
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10) )
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.3)
    ax0, ax1, ax2, ax3 = axes.flat
    # Histo for first quartile
    tmp = histog(log100(q1),bins)
    ax0.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'r', c = 'b', markersize=2,elinewidth=1.5)
    ax0.bar( tmp[0] , log100(tmp[1]), width = tmp[0][1]-tmp[0][2], alpha = 0.4, align = 'center' )
    ax0.set_title('First Quartile', fontsize = 18)
    ax0.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax0.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax0.set_ylim(min(log100(tmp[1]))-1,0)
    ax0.set_xlim(amin(tmp[0])-1/2.,amax(tmp[0])+1/2.)
    # Other quartiles
    tmp = histog(log100(q2),bins)
    ax1.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'r', c = 'b', markersize=2,elinewidth=1.5)
    ax1.bar( tmp[0], log100(tmp[1]), width = tmp[0][1]-tmp[0][2], alpha = 0.4 , align = 'center' )
    ax1.set_title('Second Quartile', fontsize = 18)
    ax1.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax1.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax1.set_ylim(min(log100(tmp[1]))-1,0)
    ax1.set_xlim(amin(tmp[0])-1/2.,amax(tmp[0])+1/2.)
    # --
    tmp = histog(log100(q3),bins)
    ax2.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'r', c = 'b', markersize=2,elinewidth=1.5)
    ax2.bar( tmp[0], log100(tmp[1]), width = tmp[0][1]-tmp[0][2], alpha = 0.4 , align = 'center' )
    ax2.set_title('Third Quartile', fontsize = 18)
    ax2.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax2.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax2.set_ylim(min(log100(tmp[1]))-1,0)
    ax2.set_xlim(amin(tmp[0])-1/2.,amax(tmp[0])+1/2.)
    # --
    tmp = histog(log100(q4),bins)
    ax3.errorbar( tmp[0], log100(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'r', c = 'b', markersize=2,elinewidth=1.5)
    ax3.bar( tmp[0], log100(tmp[1]), width = tmp[0][1]-tmp[0][2], alpha = 0.4 , align = 'center' )
    ax3.set_title('Fourth Quartile', fontsize = 18)
    ax3.set_ylabel('$Log_{10}(N)$', fontsize = 18)
    ax3.set_xlabel('$Log_{10}(M)$', fontsize = 18)
    ax3.set_ylim(min(log100(tmp[1]))-1,0)
    ax3.set_xlim(amin(tmp[0])-1/2.,amax(tmp[0])+1/2.)
```
