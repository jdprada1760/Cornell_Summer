```python
# Reads the csv
name = "finalResult.data"
table = np.loadtxt(name, skiprows = 0, delimiter = ',').T
# mass per particle
M = 8.61E+8 / 0.7
# The volume of the sample box 
vol = (40*40*40)/(0.7**3)
#M,vol
table[0] = 4/(np.pi*table[0])
len(table.T)
```

```python
# Calculates the first quartile of logsigma
lsigma = table[0]
mass = M*table[1]
q = np.percentile(lsigma,25)
# Keeps first quartile
q1 = mass[where(lsigma < q)]
# Removes fist quartile from lsigma
mass = mass[where(lsigma >= q)]
lsigma = lsigma[where(lsigma >= q)]
# Second quartile
q = np.percentile(lsigma,33.333333)
q2 = mass[where(lsigma < q)]
mass = mass[where(lsigma >= q)]
lsigma = lsigma[where(lsigma >= q)]
# Third and Fourth quartile
q = np.percentile(lsigma,50)
q3 = mass[where(lsigma < q)]
q4 = mass[where(lsigma >= q)]
# verify the length of each quartile
print len(q1),len(q2),len(q3),len(q4), sum([len(q1),len(q2),len(q3),len(q4)])
```

```python
# Gives the x,y of the histogram for logmasses from the given array of mass
def histog( mf, nbins ):
    histy = []
    histx = []
    actual = min(mf)
    dx = (max(mf) - min(mf))/nbins
    for i in range(nbins-1):
        histx.append(actual+i*dx + dx/2)
        histy.append(len(where((mf >= actual+i*dx) & (mf < actual+(i+1)*dx))[0]))
    i += 1
    histx.append(actual+i*dx + dx/2)
    histy.append(len(where((mf >= actual+i*dx) & (mf <= actual+(i+1)*dx))[0]))
    return (array(histx), array(histy)/vol)
```

```python
# To avoid log10(0)
def log100(x):
    return log10(x)
```

```python
# Schechter function
def Schechter(m,alpha,phi_s,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))*numpy.exp(-MHI/M_s)
# Power fit
def Power(m,alpha,phi_s,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))
# Returns the fitted parameterts for the Schechter function given a mass distribution
def fit(mf,bins):
    # tmp[0]:mj,  tmp[1]:nj
    tmp = histog(log100(mf),bins)
    return optimize.curve_fit(Schechter,tmp[0],tmp[1],p0=[-2.0,0,14], sigma = np.sqrt(tmp[1]), maxfev = 40000)
# Returns the fitted parameters for Power law given the mass distribution
def fitP(mf,bins):
    tmp = histog(log100(mf),bins)
    return optimize.curve_fit(Power,tmp[0],tmp[1],p0=[-2.0,0,14], sigma = np.sqrt(tmp[1]), maxfev = 40000)
```

```python
# Graphics of the mass function for each quartile
import matplotlib.gridspec as gridspec
bins = 15
fig, axes = plt.subplots( nrows=2, ncols=4, figsize=(15,20) )
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.3)
gs = gridspec.GridSpec(4,2)
ax0 = plt.subplot(gs[0, 0])
ax1 = plt.subplot(gs[0, 1])
ax2 = plt.subplot(gs[1, 0])
ax3 = plt.subplot(gs[1, 1])
ax4 = plt.subplot(gs[2:,:])
# Histo for first quartile
tmp = histog(log100(q1),bins)
logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
ax0.scatter( tmp[0], log100(tmp[1]))
ft = fit(q1,bins)[0]
ax0.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r', 
         label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"$m_\ast$ = " + str(ft[2]) , lw = 1.5)
ax0.set_title('First Quartile', fontsize = 18)
ax0.set_ylabel('$Log_{10}(N)$', fontsize = 18)
ax0.set_xlabel('$Log_{10}(M)$', fontsize = 18)
ax0.set_ylim(-6,0)
ax0.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
ax0.legend()
#--
tmp = histog(log100(q2),bins)
logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
ax1.scatter( tmp[0], log100(tmp[1]))
ft = fit(q2,bins)[0]
ax1.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
         label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"$m_\ast$ = " + str(ft[2]), lw = 1.5)
ax1.set_title('Second Quartile', fontsize = 18)
ax1.set_ylabel('$Log_{10}(N)$', fontsize = 18)
ax1.set_xlabel('$Log_{10}(M)$', fontsize = 18)
ax1.set_ylim(-6,0)
ax1.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
ax1.legend()
#--
tmp = histog(log100(q3),bins)
logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
ax2.scatter( tmp[0], log100(tmp[1]))
ft = fit(q3,bins)[0]
ax2.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
         label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"$m_\ast$ = " + str(ft[2]), lw = 1.5)
ax2.set_title('Third Quartile', fontsize = 18)
ax2.set_ylabel('$Log_{10}(N)$', fontsize = 18)
ax2.set_xlabel('$Log_{10}(M)$', fontsize = 18)
ax2.set_ylim(-6,0)
ax2.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
ax2.legend()
#--
tmp = histog(log100(q4),bins)
logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
ax3.scatter( tmp[0], log100(tmp[1]))
ft = fit(q4,bins)[0]
ax3.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r',
         label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"$m_\ast$ = " + str(ft[2]), lw = 1.5)
ax3.set_title('Fourth Quartile', fontsize = 18)
ax3.set_ylabel('$Log_{10}(N)$', fontsize = 18)
ax3.set_xlabel('$Log_{10}(M)$', fontsize = 18)
ax3.set_ylim(-6,0)
ax3.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
ax3.legend()

# Total fit
q1 = M*table[1]
bins = 20
tmp = histog(log100(q1),bins)
logm = linspace(amin(tmp[0])-0.5,amax(tmp[0])+0.5, 100)
ax4.scatter( tmp[0], log100(tmp[1]))
ft = fit(q1,bins)[0]
ax4.plot(logm, log100(Schechter(logm,ft[0],ft[1],ft[2]) ),'r', 
         label = r"$\alpha$ = " + str(ft[0]) +"\n" + r"$m_\ast$ = " + str(ft[2]) , lw = 2)
ax4.set_title('Total', fontsize = 22)
ax4.set_ylabel('$Log_{10}(N)$', fontsize = 22)
ax4.set_xlabel('$Log_{10}(M)$', fontsize = 22)
ax4.set_ylim(-6,0)
ax4.set_xlim(amin(tmp[0])-1,amax(tmp[0])+1)
ax4.legend(prop={'size':20})
```
