## Some useful tools and definitions

```python
#mass units
M = (10.0**10)/0.7
#vol of illustris run
vol = (75.0/(0.7))**3
print log(vol)

# Gives the x,y of the histogram for logmasses from the given array of mass
def histog( mf, nbins ):
    hist, bin_edges = np.histogram(mf, bins = nbins)
    # Obtains the center point for each bin
    xcenter = (bin_edges[:-1] + bin_edges[1:])/2
    return np.array(xcenter), np.array(hist/vol)

# Schechter function
def Schechter(m,phi_s,alpha,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))*numpy.exp(-MHI/M_s)

# Returns the fitted parameterts for the Schechter function given a mass distribution
def fit(tmp):
    # tmp[0]:mj,  tmp[1]:nj
    return optimize.curve_fit(Schechter,tmp[0] ,tmp[1],p0=[0,-2.0,12], sigma = np.sqrt(tmp[1]/vol),
                              #bounds = (np.array([0,-np.inf,8.]),np.array([np.inf,np.inf,20.])),method = 'trf',
                              maxfev = 100000)

# Truncates float to n decimal parts
def truncate(d,n):
    before_dec, after_dec = str(d).split('.')
    d = float('.'.join((before_dec, after_dec[0:n-1])))
    return d
```

## Loading function

```python
# loads the file with halo masses and makes the quartiles
# file formated env,mass
# j defines if we treat gass or dm j = 0,1
# minMass minimum mass allowed to have into account a halo
def loadHaloes(name, neigh, j, minMass):
    table = np.loadtxt(name, skiprows = 0, delimiter = ',').T
    #table[0] = neigh/(np.pi*table[0]+1e-9)
    print "number of haloes before: " + str(len(table.T))
    
    # Only take the haloes with not null mass
    lsigma = table[0][where((M*table[j+1]) > minMass)]
    mass = M*table[j+1][where(M*table[j+1]> minMass)]
    
    print "number of haloes after: " + str(len(mass))

    # Calculates the first quartile of logsigma
    global q1
    global q2
    global q3
    global q4
    global q5
    q5 = M*table[j+1][where(M*table[j+1]> minMass)]

    
    q = np.percentile(lsigma,25)
    # Keeps first quartile
    q1 = mass[where(lsigma <= q)]
    # Removes fist quartile from lsigma
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Second quartile
    q = np.percentile(lsigma,33.333333)
    q2 = mass[where(lsigma <= q)]
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Third and Fourth quartile
    q = np.percentile(lsigma,50)
    q3 = mass[where(lsigma <= q)]
    q4 = mass[where(lsigma >= q)]
    
    # verify the length of each quartile
    print len(q1),len(q2),len(q3),len(q4), sum([len(q1),len(q2),len(q3),len(q4)])

# Loads the file with environment classification T-Web
# file formated env,massgas,masdm
# j defines if we treat gass or dm j = 0,1
# minMass minimum mass allowed to have into account a halo
def loadTweb(name, neigh, j, minMass):
    table = np.loadtxt(name, skiprows = 0, delimiter = ',').T
    #table[0] = neigh/(np.pi*table[0]+1e-9)
    print "number of haloes before: " + str(len(table.T))
    
    # Only take the haloes with not null mass
    lsigma = table[0][where((M*table[j+1]) > minMass)]
    mass = M*table[j+1][where(M*table[j+1]> minMass)]
    
    print "number of haloes after: " + str(len(mass))

    # Calculates the first quartile of logsigma
    global q1 
    global q2 
    global q3 
    global q4 
    global q5 
    q1 = mass[np.where(lsigma == 3)]
    q2 = mass[np.where(lsigma == 2)]
    q3 = mass[np.where(lsigma == 1)]
    q4 = mass[np.where(lsigma == 0)]
    q5 = mass
    # verify the length of each quartile
    print len(q1),len(q2),len(q3),len(q4), sum([len(q1),len(q2),len(q3),len(q4)])
```

```python
# Monte Carlo Markov Chain for fitting
# n -> iterations
# sigmas -> gaussian parameter
# p0 initial guess
# data -> datax, datay
# fitfun -> fitting function
def MCMC(n,p0,sigma,data,fitfun):
    data = array(data)
    # Model variables to fit
    var = [[p0[i] for i in range(len(p0))]]
        
    # Calculates the likelihood for p0
    # likelihood defined as exp((1.0/2.0)*np.sum((real_y-model_y)**2.0)/N)
    lkhood = thehood(fitfun(data[0],var[0]),data[1],0)
    
    # Saves the likelihood for post analysis
    likelihood = [lkhood]
    
    # Last variables
    lastvar = var[0]
    
    # ITERATION 
    l = 0
    for l in range(n):
        
        # Random step
        vartemp = np.random.normal(lastvar, sigma)
        #print vartemp
                
        # If templkhood is less it makes the step, if not, it leaves it to fate
        tmplkhood = thehood(fitfun(data[0],vartemp),data[1],0)
        
        if(np.random.random() <= (tmplkhood/likelihood[-1])):
            var.append(vartemp)
            lastvar = vartemp
            likelihood.append(tmplkhood)
    print "last",l
    return var,likelihood

# Defines likelihood taking into account errors
def thehood(realdata,fitdata,errors):
    #return np.exp(-(1.0/(2.0*len(realdata)))*np.sum((exp(fitdata)-exp(realdata))**2))
    return np.exp(-(1.0/(2.0*len(realdata)))*np.sum((fitdata-realdata)**2))

# SuperMCMC
def SuperMCMC(n,p0,sigma,data,fitfun):
    var,lkhood = MCMC(n,p0,sigma,data,fitfun)
    newp0 = var[argmax(lkhood)]
    var2,lkhood2 = MCMC(n/2,newp0,sigma/4.,data,fitfun)
    return var2+var,lkhood2+lkhood
    

# Schechter function
def SchechterMC(m,params):
    MHI = 10.**m
    M_s = 10.**params[2]
    return np.log(10.)*params[0]*((MHI/M_s)**(params[1]+1))*numpy.exp(-MHI/M_s)
```
