## Some useful tools and definitions

```python
#mass per particle of gas
Mg = 8.85678e+5 /0.7
#mass per particle of dm
Mdm = 0.000440896524361e+10/0.7
#vector of mass
M = [Mg,Mdm]
#vol of illustris run
vol = (75e+3/0.7)**3

# Gives the x,y of the histogram for logmasses from the given array of mass
def histog( mf, nbins ):
    hist, bin_edges = np.histogram(mf, bins = nbins)
    # Obtains the center point for each bin
    xcenter = (bin_edges[:-1] + bin_edges[1:])/2
    return np.array(xcenter), np.array(hist/vol)

# Schechter function
def Schechter(m,alpha,phi_s,m_s):
    MHI = 10.**m
    M_s = 10.**m_s
    return np.log(10.)*phi_s*((MHI/M_s)**(alpha+1))*numpy.exp(-MHI/M_s)

# Returns the fitted parameterts for the Schechter function given a mass distribution
def fit(mf,bins):
    # tmp[0]:mj,  tmp[1]:nj
    tmp = histog(log100(mf),bins)
    return optimize.curve_fit(Schechter,tmp[0] ,tmp[1],p0=[-2.0,0,12], sigma = np.sqrt(tmp[1]*vol)/vol, maxfev = 100000)
```

## Loading function

```python
# loads the file with halo masses and makes the quartiles
# file formated env,mass
# j defines if we treat gass or dm
# minMass minimum mass allowed to have into account a halo
def loadHaloes(name, neigh, j, minMass):
    table = np.loadtxt(name, skiprows = 0, delimiter = ',').T
    # adds 1e-9 to avoid possible divisions by 0
    table[0] = neigh/(np.pi*table[0])
    print "number of haloes before: " + str(len(table.T))
    
    # Only take the haloes with not null mass
    lsigma = table[0][where((M[j]*table[1]) > minMass)]
    mass = M[j]*table[1][where((M[j]*table[1])> minMass)]
    
    print "number of haloes after: " + str(len(mass))

    # Calculates the first quartile of logsigma
    global q1
    global q2
    global q3
    global q4
    global q5
    q5 = M[j]*table[1][where(M[j]*table[1]> minMass)]

    
    q = np.percentile(lsigma,25)
    # Keeps first quartile
    q1 = mass[where(lsigma <= q)]
    # Removes fist quartile from lsigma
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Second quartile
    q = np.percentile(lsigma,33.333333)
    q2 = mass[where(lsigma <= q)]
    mass = mass[where(lsigma > q)]
    lsigma = lsigma[where(lsigma > q)]
    # Third and Fourth quartile
    q = np.percentile(lsigma,50)
    q3 = mass[where(lsigma <= q)]
    q4 = mass[where(lsigma >= q)]
    
    # verify the length of each quartile
    print len(q1),len(q2),len(q3),len(q4), sum([len(q1),len(q2),len(q3),len(q4)])
```

## Histogram graphing function (can be shorter ithink)

```python
def histo(bins):
    # Graphics of the mass function for each quartile
    import matplotlib.gridspec as gridspec
    fig, axes = plt.subplots( nrows=2, ncols=4, figsize=(15,20) )
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.3)
    gs = gridspec.GridSpec(4,2)
    ax0 = plt.subplot(gs[0, 0])
    ax1 = plt.subplot(gs[0, 1])
    ax2 = plt.subplot(gs[1, 0])
    ax3 = plt.subplot(gs[1, 1])
    ax4 = plt.subplot(gs[2:,:])
    
    # Callable lists
    axs = [ax0,ax1,ax2,ax3,ax4]
    qs = [q1,q2,q3,q4,q5]
    indx = ['1st quartile','2nd quartile','3rd quartile','4th quartile', 'Total']
    
    for ax,q,i in zip(axs,qs,indx):
        # Histo for first quartile
        tmp = histog(log10(q),bins)
        ax.errorbar( tmp[0], log10(tmp[1]), yerr = sqrt(vol*tmp[1])/(vol*tmp[1]), fmt='o', ecolor = 'r', c = 'b', markersize=3,elinewidth=1.5)
        #ax.bar( tmp[0] , log10(tmp[1]), width = tmp[0][1]-tmp[0][2], alpha = 0.4, align = 'center' )
        ax.set_title(i, fontsize = 18)
        ax.set_ylabel('$Log_{10}(N)$', fontsize = 18)
        ax.set_xlabel('$Log_{10}(M)$', fontsize = 18)
        ax.set_xlim(min(tmp[0])-1,max(tmp[0])+1)
        ax.set_ylim(min(log10(tmp[1]))-1,(max(log10(tmp[1]))+1))
    
```
