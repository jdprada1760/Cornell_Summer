```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_summary(filename):
    dtype=[('gas', 'f8'),
           ('dm', 'f8'), 
           ('star', 'f8'),
           ('bh', 'f8'),
           ('2d', 'f8'),
           ('3d', 'f8')]
    summary = np.loadtxt(filename, dtype=dtype, delimiter=',')    
    return summary
```

```python
data = load_summary("../../data/nnEnv/specs1.data")
```

```python
plt.scatter((data['2d']), (data['3d']), s=0.1)
plt.xlabel('2d')
plt.ylabel('3d')
```

```python
plt.scatter(np.log10(data['gas']), (data['3d']), s=0.1)
plt.xlabel('log_10 gas')
plt.ylabel('2d')
```
